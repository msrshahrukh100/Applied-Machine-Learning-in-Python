__Import necessary modules__

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('fruit_data_with_colors.txt', sep='\t')
```

```python
df.head()
```

__The features X and labels y are separated here__

```python
X = df.loc[:, ['fruit_label','mass', 'width', 'height', 'color_score']]
y = df['fruit_label']
X.head()
```

```python
sns.pairplot(X, hue='fruit_label')
plt.show()
```

```python
# plotting a 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X['width'], X['height'], X['color_score'], c = y, marker = 'o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```




*Here we'll be using KNN algorithm for Classification. KNN is a instance based or memory based as it memorizes the labelled examples in the trainig set and uses for classifications. KNN can be used for classification as well as regression problems. It has 3 steps* - 
1. Find k most similar instances based on distance basically
2. Get their labels
3. Decide the label of the unknown sample using simple voting.
A few parameters needs to be set -
1. How the relatedness is to be measured. (Eucledian distance here)
2. How many neighbors to consider.
3. How the label is decided. (Majority voting here)

### Applying KNN on the dataset. 





```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X.drop('fruit_label', axis=1, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
y_train.shape
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

*Here the accuracy seems to be less. Now we'll use different values for n_neighbors to see which ones gives the best values. We can also use cross validation for that. But here we'll proceed in the simpler way*

```python
scores = []
neighbors_values = range(1,11)

for n in neighbors_values :
    knn = KNeighborsClassifier(n_neighbors = n)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.clf()
plt.plot(neighbors_values, scores)
plt.show()
```

__Import necessary modules__

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('fruit_data_with_colors.txt', sep='\t')
```

```python
df.head()
```

__The features X and labels y are separated here__

```python
X = df.loc[:, ['fruit_name', 'fruit_subtype', 'mass', 'width', 'height', 'color_score']]
y = df.loc[:,['fruit_label']]
```

```python
X.head()
```

```python
sns.pairplot(X, hue='fruit_name')
```

```python
plt.show()
```

```python
# plotting a 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X['width'], X['height'], X['color_score'], c = y, marker = 'o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```
